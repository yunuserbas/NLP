```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk

text = """Pelé, the Brazilian king of soccer who won a record 3 World Cups and became one of the most commanding sports figures of the last century, died Thursday. 
He was 82. Many of those goals became legendary, but Pelé’s influence on the sport went well beyond scoring. 
He helped create and promote what he later called “o jogo bonito” — the beautiful game — a style that valued clever ball control, 
inventive pinpoint passing and a voracious appetite for attacking. Pelé not only played it better than anyone; he also championed it around the world.
Among his athletic assets was a remarkable center of gravity; as he ran, swerved, sprinted or backpedaled, his midriff seemed never to move, 
while his hips and his upper body swiveled around it.
He could accelerate, decelerate or pivot in a flash. Off-balance or not, he could lash the ball accurately with either foot. 
Relatively small, at 5 feet 8 inches, he could nevertheless leap exceptionally high, often seeming to hang in the air to put power behind a header. 
Like other sports, soccer has evolved. Today, many of its stars can execute acrobatic shots or rapid-fire passing sequences. But in his day, 
Pelé’s playmaking and scoring skills were stunning."""   
```

**1- generate_from_text**

A WordCloud from the initial text, using the generate_from_text method

```python
wordcloud = WordCloud(background_color='white', max_words=50).generate_from_text(text)
```

```python
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

**2- generate_from_frequencies**

A WordCloud using the generate_from_frequencies method after cleaning up stopwords and punctuation.

```python
words = nltk.word_tokenize(text)
```

```python
import string 

# Remove the stopwords and punctuation
filtered_words = [word for word in words if word.lower() not in nltk.corpus.stopwords.words('english') and word not in string.punctuation]
```

```python
word_frequencies = {}

for word in filtered_words:
    if word in word_frequencies:
        word_frequencies[word] += 1
    else:
        word_frequencies[word] = 1
```

```python
wordcloud = WordCloud().generate_from_frequencies(word_frequencies)

plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

**3- generate_from_frequencies**

A WordCloud thanks to the generate_from_frequencies method after cleaning up stopwords and punctuation, and applying a stemmer.

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
```

```python
word_frequencies2 = {}

# Iterate through the list of words and update the word frequencies
for word in stemmed_words:
    if word in word_frequencies2:
        word_frequencies2[word] += 1
    else:
        word_frequencies2[word] = 1
```

```python
wordcloud3 = WordCloud().generate_from_frequencies(word_frequencies2)
```

```python
plt.figure()
plt.imshow(wordcloud3, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

**4- generate_from_frequencies**

A WordCloud using the generate_from_frequencies method after cleaning up stopwords and punctuation, and applying a lemmatizer.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
```

```python
word_frequencies3 = {}

# Iterate through the list of words and update the word frequencies
for word in lemmatized_words:
    if word in word_frequencies3:
        word_frequencies3[word] += 1
    else:
        word_frequencies3[word] = 1
```

```python
wordcloud4 = WordCloud().generate_from_frequencies(word_frequencies3)

# Show the word cloud
plt.figure()
plt.imshow(wordcloud4, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
